[Pandas read data from excel](https://pandas.pydata.org/docs/reference/api/pandas.read_excel.html)

```python
df = pd.read_excel(datapath)
```

```python
df
```

```python
df.columns[0]
```

```python
row_indexs = df.iloc[:,0]
```

[Stackoverflow: To drop a column](https://stackoverflow.com/questions/13411544/delete-a-column-from-a-pandas-dataframe)

```python
df.drop(df.columns[0], axis=1, inplace=True)
df
```

```python
import matplotlib.pyplot as plt
```

[Pandas plot dataframe](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.html)

[stackoverflow: How incorporate styles for different lines in the plot](https://stackoverflow.com/questions/43941245/line-plot-with-data-points-in-pandas)

[stackoverflow: Dataframe plot marker styles](https://stackoverflow.com/questions/8409095/set-markers-for-individual-points-on-a-line-in-matplotlib)



```python
plt.rcParams['figure.figsize'] = [10, 6]
plt.rcParams['figure.dpi'] = 100 # 200 e.g. is really fine, but slower

df.plot(style=["d-", ">-", "^-", "P--","s:", "p:", "*-", "D-"],
       xlabel='Number of Top Matches Considered',
       ylabel="ARP%")

plt.grid()
plt.xticks()
plt.show()
```
